```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist

(x_train, y_train), (_, _) = mnist.load_data()
x_train = x_train.astype('float32') / 255.0

image = x_train[0].reshape(1, 28, 28, 1)  

filter_weights = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]], dtype=np.float32).reshape(3, 3, 1, 1)
filter_tensor = tf.constant(filter_weights)

feature_map = tf.nn.conv2d(image, filter_tensor, strides=[1, 1, 1, 1], padding='VALID')
feature_map = feature_map.numpy().squeeze()

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image.squeeze(), cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Feature Map Output")
plt.imshow(feature_map, cmap='gray')

plt.show()

```
