```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

image = np.random.rand(64, 64).astype(np.float32) 

image = image.reshape(1, 64, 64, 1) 

filter_weights = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
], dtype=np.float32).reshape(3, 3, 1, 1)

image_tensor = tf.constant(image)
filter_tensor = tf.constant(filter_weights)

feature_map = tf.nn.conv2d(image_tensor, filters=filter_tensor, strides=1, padding="VALID")

feature_map_np = feature_map.numpy().squeeze()

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image.squeeze(), cmap="gray")
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(feature_map_np, cmap="gray")
axes[1].set_title("Feature Map (After Convolution)")
axes[1].axis("off")

plt.show()

print("Feature Map Shape:", feature_map_np.shape)
```
