```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

feature_map = np.random.randn(62, 62).astype(np.float32) 

feature_map = feature_map.reshape(1, 62, 62, 1)

relu_feature_map = tf.nn.relu(feature_map)

pooled_feature_map = tf.nn.max_pool(relu_feature_map, ksize=2, strides=2, padding="VALID")

relu_feature_map_np = relu_feature_map.numpy().squeeze()
pooled_feature_map_np = pooled_feature_map.numpy().squeeze() 

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(feature_map.squeeze(), cmap="gray")
axes[0].set_title("Original Feature Map (62x62)")
axes[0].axis("off")

axes[1].imshow(relu_feature_map_np, cmap="gray")
axes[1].set_title("After ReLU (62x62)")
axes[1].axis("off")

axes[2].imshow(pooled_feature_map_np, cmap="gray")
axes[2].set_title("After Max Pooling (31x31)")
axes[2].axis("off")

plt.show()

print("Feature Map Shape after ReLU:", relu_feature_map_np.shape)
print("Feature Map Shape after Max Pooling:", pooled_feature_map_np.shape)
```
